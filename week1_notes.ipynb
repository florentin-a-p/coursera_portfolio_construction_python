2 ways of thinking about return or **price return**:

$$\dfrac{P_{t+1}-P_{t}}{P_{t}}$$
or the 1+r format
$$\dfrac{P_{t+1}}{P_{t}}$$
**Total return** is when you add dividend to the return or price return
$$R_{t,t+1}=\dfrac{P_{t+1}-P_{t}}{P_{t}}+\dfrac{D_{t,t+1}}{P_{t}}=\dfrac{P_{t+1}-P_{t}+D_{t,t+1}}{P_{t}}=\dfrac{P_{t+1}+D_{t,t+1}}{P_{t}}-1$$
Total return - Price return = **contribution of dividends**
$$\dfrac{P_{t+1}+D_{t,t+1}-P_{t}}{P_{t}}-\dfrac{P_{t+1}-P_{t}}{P_{t}} = \dfrac{P_{t+1}+D_{t,t+1}-P_{t}-P_{t+1}+P_{t}}{P_{t}} = \dfrac{D_{t,t+1}}{P_{t}}$$ <br/>

**Compounding returns**
$$[\Pi_{i=0}^{n}(1+R_{t_{i},t_{i+1}})]-1$$
you buy a stock that returns 10% on 1st day, then returns -3% on 2nd day
$$(1+.10)(1-.03)-1 == 0.067 = 6.7%$$

**Annualizing returns** <br/>
let's say the return per a period (e.g. month) is fixed at 1%, it means we can simplify the following
$$[\Pi_{i=0}^{n}(1+R_{t_{i},t_{i+1}})]-1$$
into 
$$[(1+R_{t_{i},t_{i+1}})^n]-1$$

**Volatility**

The blue series is much less volatile than the yellow series. The average of these 2 series is exactly the same, but the yellow moves from the mean much more often.
<div>

</div>

in statistics, 

$$\text{volatility}=\text{standard deviation}$$

Standard deviation is defined as following
$$\sigma = \sqrt{\frac{1}{N}\sum_{i=1}^N(R_i-\bar{R})^2}$$
where $\bar{R}$ is the arithmetic mean of the returns <br/>
the yellow series has much higher variance than the blue series


**Annualizing volatility**
$$\sigma_{\text{annualized}} = \sigma_p\sqrt{p}$$

to get the annualized volatility, you multiply the volatility of a certain period with the number of period within a year <br/>
to get the annualized volatility, you multiply the daily volatility with the number of days per year <br/>
remember the number of days per year is 252, not 365 <br/>
annualized volatility is always greater than monthly volatility
